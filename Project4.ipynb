```python
# 1. Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset

df = pd.read_csv("train.csv")

# 3. Data Inspection

print("\n========== DATA INFO ==========")
df.info()

print("\n=========== Shape (Rows, Columns) ============")
print(df.shape)

print("\n========== STATISTICAL SUMMARY ==========")
print(df.describe(include='all'))

print("\n========== FIRST 5 ROWS ==========")
print(df.head())

print("\n========== LAST 5 ROWS ==========")
print(df.tail())

print("\n========== Number of Rows ==========")
print(len(df))

print("\n========== Column Names ===========")
print(df.columns.tolist())

# 4. Data Cleaning

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

print("\n========== DUPLICATE ROWS ==========")
print(df.duplicated().sum())

# ANALYTICAL QUESTIONS

# Q1. What is the overall placement rate of students?

placement_counts = df['Placement_Status'].value_counts()
print("\n",placement_counts)
placement_rate = (df['Placement_Status'] == 'Placed').mean() * 100
print(f"Overall Placement Rate: {placement_rate:.2f}")
print("Insight:")
print("Only 36.2% of students were placed, while the majority (63.8%) were not placed.")

# Q2. Does CGPA affect placement status?

print("\n",df.groupby('Placement_Status')['CGPA'].mean())
plt.figure(figsize=(8,5))
sns.boxplot(x='Placement_Status', y='CGPA', data=df)
plt.title('CGPA vs Placement Status')
plt.xlabel('Placement Status')
plt.ylabel('CGPA')
plt.show()
print("Insight:")
print("Placed students have a noticeably higher CGPA than non-placed students, indicating CGPA positively affects placement.")

# Q3. What is the gender distribution of students?

gender_count = df['Gender'].value_counts()
print("\n",gender_count)
gender_count.plot(kind='bar')
plt.title("Gender Distribution of Students")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.show()
print("Insight:")
print("The dataset has an almost equal gender distribution (2260 females and 2238 males).")

# Q4. How many students have completed at least one internship?

stud_1_intern = (df['Internships'] >= 1).sum()
print("\nNumber of students who completed at least one internship:", stud_1_intern)
print("Insight:")
print("24365 internship records show that morre than half of the students have gained practical industry experience.")

# Q5. Which branch has the highest number of students?

print("\n",df['Branch'].value_counts().idxmax())
print("Insight:")
print("ME (Mechanical Engineering) branch has the highest number os students in the dataset.")

# Q6. What is the average CGPA of all students?

average_cgpa = df['CGPA'].mean()
print("\nAverage CGPA of all students:", average_cgpa)
print("Insight:")
print("The average CGPA of students is 7.00, indicating overall good academic performance.")

# Q7. What percentage of students are placed and not placed?

placement_counts = df['Placement_Status'].value_counts()
plt.pie(placement_counts,labels=placement_counts.index,autopct='%1.1f%%')
plt.title('\nPlacement Status Distribution')
plt.show()
print("Insight:")
print("36.2% of students are placed, whereas 63.8% are not placed, showing placement opportunities can be improved.")

print("\n================== Conclusion ======================")
print("This project analyzed student placement data to understand the factors affecting placements. The")
print("analysis showed that students with higher CGPA and internship experience have better placement")
print("opportunities.The dataset also provided insights into gender distribution, branch-wise student count,")
print("and overall academic performance. Overall, this analysis help in understanding student placement trends")
print("and the importance of academics and practical experience.")
```
