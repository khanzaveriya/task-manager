```python
# 1. Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset

df = pd.read_csv("hospital data analysis.csv")

# 3. Data Inspection

print("\n========== DATA INFO ==========")
df.info()

print("\n=========== Shape (Rows, Columns) ============")
print(df.shape)

print("\n========== STATISTICAL SUMMARY ==========")
print(df.describe(include='all'))

print("\n========== FIRST 5 ROWS ==========")
print(df.head())

print("\n========== LAST 5 ROWS ==========")
print(df.tail())

print("\n========== Number of Rows ==========")
print(len(df))

print("\n========== Column Names ===========")
print(df.columns.tolist())

# 4. Data Cleaning

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

print("\n========== DUPLICATE ROWS ==========")
print(df.duplicated().sum())

# ANALYTICAL QUESTIONS

# Q1. How many patients gave each satisfaction score?

satisfaction_counts = df["Satisfaction"].value_counts().sort_index()
print("\nNumber of Patients for Each Satisfaction Score:")
print(satisfaction_counts)
print("Insight:")
print("The most common satisfaction score was 4, given by 458 patients.")

# Q2. Which procedures are performed most frequently?

procedure_count = df["Procedure"].value_counts()
print("\nProcedure Frequency:")
print(procedure_count)
procedure_count.plot(kind="bar")
plt.title("Frequency of Procedures")
plt.xlabel("Procedure")
plt.ylabel("Number of Patients")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
print("Insight:")
print("The most frequently performed procedure was Cast and Physical Therapy, indicating")
print("the highest patient demand for this treatment.")


# Q3. What is the most common medical condition among patients?

condition_count = df["Condition"].value_counts()
most_common_condition = condition_count.idxmax()
patient_count = condition_count.max()
print("\nPatient Count by Medical Condition:")
print(condition_count)
print("\nMost Common Medical Condition:")
print(f"{most_common_condition} -> {patient_count} patients")
print("Insight:")
print("The most common medical condition among patients is Fractured Leg, affecting 67 patients.")

# Q4. Which gender has visited the hospital more frequently?

gender_count = df["Gender"].value_counts()
top_gender = gender_count.idxmax()
visit_count = gender_count.max()
print("\nPatient Count by Gender:")
print(gender_count)
print("\nGender with Highest Hospital Visits:")
print(f"{top_gender} -> {visit_count} visits")
print("Insight:")
print("The hospital received the highest number of visits from female patients 524 visits.")

# Q5. What percentage of patients were readmitted?

readmitted_count = (df["Readmission"] == "Yes").sum()
total_patients = len(df)
readmission_percentage = (readmitted_count / total_patients) * 100
print(f"\nReadmitted Patients: {readmitted_count}")
print(f"Total Patients: {total_patients}")
print(f"Readmission Percentage: {readmission_percentage:.2f}%")
print("Insight:")
print("Approximately 26.83% of patients were readmitted to the hospital.")

# Q6. What is th age distribution of patients?

plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=10)
plt.title("\nAge Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
print("Insight:")
print("Most patients belong to the age group of 50-60 years.")

print("\n======================= Conclusion ======================")
print("We have successfully performed data cleaning, analysis, and")
print("visualization on the hospital dataset, uncovering key insights")
print("that can support better healthcare decision-making.")
```
